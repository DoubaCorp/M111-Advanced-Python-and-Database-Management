## Exercise 1
**Read Total profit of all months and show it using a line plot**

Total profit data provided for each month. Generated line plot must include the following properties:
- X label name = Month Number
- Y label name = Total profit
The line plot graph should look like this.

**Expected output**

<img src="../data/Company_profit_per_month.png" alt="Company_profit_per_month" style="width: 500px;"/>


```python
import pandas as pd
import matplotlib.pyplot as plt  

df = pd.read_csv("../data/company_sales_data.csv")
p = df ['total_profit'].tolist()
m = df ['month_number'].tolist()
plt.plot(m, p, label = 'Month-wise Profit data of last year')
plt.xlabel('Month number')
plt.ylabel('Profit in dollar')
plt.xticks(m)
plt.title('Company profit per month')
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.show()
```

## Exercise 2
**Get total profit of all months and show line plot with the following Style properties**
- Line Style dotted and Line-color should be red
- Show legend at the lower right location.
- X label name = Month Number
- Y label name = Sold units number
- Add a circle marker.
- Line marker color as read
- Line width should be 3

**Expected output**

<img src="../data/company_sales_data_of_last_year.png" alt="company_sales_data_of_last_year" style="width: 700px;"/>


```python
df = pd.read_csv("../data/company_sales_data.csv")
pt = df ['total_profit'].tolist()
m = df ['month_number'].tolist()

plt.plot(m,p, label ='Profit data of last year', 
      color='r', marker='o', markerfacecolor='k', 
      linestyle='--', linewidth=3)
      
plt.xlabel('Month Number')
plt.ylabel('Profit in dollar')
plt.legend(loc='lower right')
plt.title('Company Sales data of last year')
plt.xticks(m)
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.show()
```

## Exercise 3
**Read all product sales data and show it  using a multiline plot**
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).
**Expected output**

<img src="../data/sales_data.png" alt="sales_data" style="width: 500px;"/>


```python
df = pd.read_csv("../data/company_sales_data.csv")
m = df ['month_number'].tolist()
fc = df ['facecream'].tolist()
fw = df ['facewash'].tolist()
tp = df ['toothpaste'].tolist()
bg = df ['bathingsoap'].tolist()
shamp = df ['shampoo'].tolist()
moist = df ['moisturizer'].tolist()

plt.plot(m, fc, label = 'Face cream Sales Data', marker='o', linewidth=1)
plt.plot(m, fw, label = 'Face Wash Sales Data',  marker='o', linewidth=1)
plt.plot(m, tp, label = 'ToothPaste Sales Data', marker='o', linewidth=1)
plt.plot(m, bg, label = 'ToothPaste Sales Data', marker='o', linewidth=1)
plt.plot(m, shamp, label = 'ToothPaste Sales Data', marker='o', linewidth=1)
plt.plot(m, moist, label = 'ToothPaste Sales Data', marker='o', linewidth=1)

plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.xticks(m)
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
plt.title('Sales data')
plt.show()
```

## Exercise 4
**Read toothpaste sales data of each month and show it using a scatter plot**
Also, add a grid in the plot. gridline style should “–“.

**Expected output**

<img src="../data/Tooth_paste_Sales_data.png" alt="sales_data" style="width: 700px;"/>

## Exercise 5
**Read face cream and facewash product sales data and show it using the bar chart**
The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

**Expected output**

<img src="../data/Facewash_and_facecream_sales_data.png" alt="sales_data" style="width: 700px;"/>


```python
df = pd.read_csv("../data/company_sales_data.csv")
m = df ['month_number'].tolist()
fc = df ['facecream'].tolist()
fw = df ['facewash'].tolist()

plt.bar([a-0.25 for a in m], fc, width= 0.25, label = 'Face Cream sales data', align='edge')
plt.bar([a+0.25 for a in m], fw, width= -0.25, label = 'Face Wash sales data', align='edge')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.title(' Sales data')

plt.xticks(m)
plt.grid(True, linewidth= 1, linestyle="--")
plt.title('Facewash and facecream sales data')
plt.show()
```

## Exercise 7
**Read the total profit of each month and show it using the histogram to see the most common profit ranges**

**Expected output**

<img src="../data/Profit_data.png" alt="Profit_data" style="width: 700px;"/>


```python
df = pd.read_csv("../data/company_sales_data.csv")
profit = df ['total_profit'].tolist()
labels = ['low', 'average', 'Good', 'Best']
pro = [150000, 175000, 200000, 225000, 250000, 300000, 350000]
plt.hist(profit, pro, label = 'Profit data')
plt.xlabel('profit range in dollar')
plt.ylabel('Actual Profit in dollar')
plt.legend(loc='upper left')
plt.xticks(pro)
plt.title('Profit data')
plt.show()
```

## Exercise 8
**Calculate total sale data for last year for each product and show it using a Pie chart**

*Note: In Pie chart display Number of units sold per year for each product in percentage.*

**Expected output**

<img src="../data/sales_data_pi_chart.png" alt="sales_data_pi_chart" style="width: 700px;"/>


```python
df = pd.read_csv("../data/company_sales_data.csv")
m = df ['month_number'].tolist()

labels = ['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
data = [df ['facecream'].sum(), df ['facewash'].sum(), df ['toothpaste'].sum(), 
         df ['bathingsoap'].sum(), df ['shampoo'].sum(), df ['moisturizer'].sum()]
plt.axis("equal")
plt.pie(data, labels=labels, autopct='%1.1f%%')
plt.legend(loc='lower right')
plt.title('Sales data')
plt.show()
```

## Exercise 9
**Read Bathing soap facewash of all months and display it using the Subplot**

**Expected output**

<img src="../data/Sales_data_of_a_facewash.png" alt="Sales_data_of_a_facewash" style="width: 700px;"/>


```python
df = pd.read_csv("../data/company_sales_data.csv")
m = df  ['month_number'].tolist()
bg = df ['bathingsoap'].tolist()
fw = df ['facewash'].tolist()

f, axarr = plt.subplots(2, sharex=True)
axarr[0].plot(m, bg, label = 'Bathingsoap Sales Data', color='k', marker='o', linewidth=3)
axarr[0].set_title('Sales data of  a Bathingsoap')
axarr[1].plot(m, fw, label = 'Face Wash Sales Data', color='r', marker='o', linewidth=3)
axarr[1].set_title('Sales data of  a facewash')

plt.xticks(m)
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.show()
```
